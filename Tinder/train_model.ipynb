```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import torchvision
import torch
import PIL
from tensorflow.keras.layers import ZeroPadding2D, MaxPooling2D, Convolution2D, Dense, Activation, Flatten, Dropout
from tensorflow.keras.models import Sequential, Model
from tensorflow import keras
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
train_df['purpose'] = 'train'
test_df['purpose'] = 'test'
```

```python
np.random.seed(17)
val_set_size = int((40*train_df.shape[0])/100)
val_idx = np.random.choice(train_df.shape[0], size=val_set_size)
train_df.loc[val_idx, 'purpose'] = 'validation'
df = pd.concat([train_df, test_df])
del train_df, test_df
```

```python
p = torchvision.transforms.Compose(
    [torchvision.transforms.Resize((224, 224)), torchvision.transforms.ToTensor()])
count = 0
```

```python
def retrievePixels(img_name):
    global count
    path = f"/kaggle/input/auto-tinder/Images/{img_name}"
    img = PIL.Image.open(path)
    img = p(img)
    img = img.numpy()
    x = img.reshape(1, -1)[0]
    print(count)
    count = count+1

    return x
```

```python
df['pixels'] = df['image'].apply(retrievePixels)
features = []
pixels = df['pixels'].values
for i in range(0, pixels.shape[0]):
    features.append(pixels[i])

features = np.array(features)
features = features.reshape(features.shape[0], 224, 224, 3)
features = features / 255  # normalize inputs within [0, 1]
```

```python
from keras.applications import VGG16
```

```python
from keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, GlobalAveragePooling2D, MaxPooling2D, GlobalMaxPooling2D
```

```python
base_model = VGG16(weights= None, include_top=False, input_shape= (224,224,3))
```

```python
num_classes=1
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
predictions = Dense(num_classes, activation= 'relu')(x)
```

```python
df['score'].max()
```

```python
model = Model(inputs = base_model.input, outputs = predictions)###Thsi is used to fit.
```

```python
from keras.optimizers import SGD, Adam
adam = Adam(lr=0.001)
model.compile(optimizer= adam, loss='mean_squared_error', metrics=['accuracy'])
```

```python
checkpointer = ModelCheckpoint(
    filepath='beauty_model.hdf5', monitor="val_loss", verbose=1, save_best_only=True, mode='auto'
)
```

```python
earlyStop = EarlyStopping(monitor='val_loss', patience=5)
train_idx = df[(df['purpose'] == 'train')].index
val_idx = df[(df['purpose'] == 'validation')].index
test_idx = df[(df['purpose'] == 'test')].index
```

```python
score = model.fit(
    features[train_idx], df.iloc[train_idx].score, epochs=5,batch_size=32, validation_data=(features[val_idx], df.iloc[val_idx].score), callbacks=[checkpointer, earlyStop]
)
```

```python
model.load_weights("beauty_model.hdf5")
model.save("model_.h5")
```
